```python
import pandas as pd
import numpy as np
from ipyleaflet import Map, Heatmap
from random import uniform

import time



```

```python
## Loading data

#gpspath = '/Users/paskin/Documents/GitHub/MTA-subway/NYC_Transit_Subway_Entrance_And_Exit_Data.csv'
stationtotals = '/Users/paskin/Documents/GitHub/MTA-subway/MTA_StationDataTotals.csv'
#stationlist = 'http://web.mta.info/developers/data/nyct/subway/Stations.csv'
#stationids = '/Users/paskin/Documents/GitHub/MTA-subway/data/MTA_StationNames.csv'
gpspath = '/Users/paskin/Documents/GitHub/MTA-subway/data/GPS_and_Address.csv'

dfgps = pd.read_csv(gpspath)
dftotals = pd.read_csv(stationtotals)

dftotals.columns = ['STATION', 'ENTRIES']
dfgps = dfgps.rename(columns={'input_string': 'Address'})



```

```python
dftotals['Address'] = dftotals['STATION'] + ' New York City metro station'

#dftotals.to_csv('Address.csv')


```

```python
## Combining datasets by name

comb = pd.merge(dftotals, dfgps, how='left', on='Address')



comb


```

```python
len(comb['ENTRIES'])
```

```python
from ipyleaflet import Map, Heatmap
from random import uniform

m = Map(center=(40.765564, -73.980331), zoom=12,  dragging = True, scroll_wheel_zoom=True, box_zoom=False)

heatmap = Heatmap(
    locations=[[comb['latitude'].tolist()[i], 
    comb['longitude'].tolist()[i], 
    comb['ENTRIES'].tolist()[0]] 
    for i in range(353)] ,
    radius=5,
    blur = 7,
    max = comb['ENTRIES'].min()
   
    
    
)

m.add_layer(heatmap);

m
```
